### Create final_score column

```python
df2_ball["final_score"] = 0

final_score = df2_ball.groupby(['id','inning']).total_runs.sum().reset_index()

for i in range(len(final_score)):
    temp = final_score.loc[i]
    df2_ball.loc[(df2_ball.id == temp.id) & (df2_ball.inning == temp.inning), "final_score"] = temp.total_runs
```

```python
df2_ball.head()
```

### Total runs, wickets and balls per over

```python
df2_ball.tail()
```

```python
# is_wicket indicates wicket in particular over, we will create sum of total_wickets till particular over
# total_runs we will create sum of runs scored in particular over.
# ball we will create sum of balls delivered in particular over.

# Create new column for wickets and runs with initial value as 0
df2_ball['wickets'] = 0
df2_ball['runs'] = 0

# for ball we will start initial value with first ball delivered i.e 1
df2_ball['ball'] = 1

id = df2_ball.id.unique()

for i in id:
    for inning in range(1,3):
        index = df2_ball.loc[(df2_ball.id == i) & (df2_ball.inning == inning)].index
        for j in range(len(index)):
            current_index_df = df2_ball.loc[(df2_ball.index == index[j])]
            previous_index_df = df2_ball.loc[(df2_ball.index == index[j-1])]
            if j == 0:
                df2_ball.loc[df2_ball.index == index[j], "wickets"] = current_index_df.is_wicket.values[0]
                df2_ball.loc[df2_ball.index == index[j], "runs"] = current_index_df.total_runs.values[0]
                df2_ball.loc[df2_ball.index == index[j], "ball"] = current_index_df.ball.values[0]
            else:
                df2_ball.loc[df2_ball.index == index[j], "wickets"] = current_index_df.is_wicket.values[0] + previous_index_df.wickets.values[0]
                df2_ball.loc[df2_ball.index == index[j], "runs"] = current_index_df.total_runs.values[0] + previous_index_df.runs.values[0]
                df2_ball.loc[df2_ball.index == index[j], "ball"] = current_index_df.ball.values[0] + previous_index_df.ball.values[0]
```

### Last 5 over's (runs, wickets, balls)

```python
df2_ball.loc[df2_ball.over < int(str(current_index_df.over.values[0]).split(".")[0]) - 5]
```

```python
# let's create new column for last_5_over_wickets
df2_ball["last_5_over_wickets"] = 0
# let's create new column for last_5_over_runs
df2_ball["last_5_over_runs"] = 0
# let's create new column for last_5_over_balls
df2_ball["last_5_over_balls"] = 0
 # calculating last run, wickets and balls of last 5 over
for i in id:
    for inning in range(1, 3):
        index = df2_ball.loc[(df2_ball.id == i) & (df2_ball.inning == inning)].index
        for j in range(len(index)):
            current_index_df = df2_ball.loc[(df2_ball.index == index[j])]
            previous_index_df = df2_ball.loc[(df2_ball.index == index[j-1])]            
            if current_index_df.over.values[0] <= 5:
                df2_ball.loc[df2_ball.index == index[j], "last_5_over_wickets"] = current_index_df.wickets.values[0]
                df2_ball.loc[df2_ball.index == index[j], "last_5_over_runs"] = current_index_df.runs.values[0]
                df2_ball.loc[df2_ball.index == index[j], "last_5_over_balls"] = current_index_df.ball.values[0]
            else:
                df2_ball.loc[df2_ball.index == index[j], "last_5_over_wickets"] = current_index_df.wickets.values[0] - df2_ball.loc[(df2_ball.id == i) & (df2_ball.inning == inning) & (df2_ball.over < int(str(current_index_df.over.values[0]).split(".")[0]) - 5+1), "wickets"].values[0]
                df2_ball.loc[df2_ball.index == index[j], "last_5_over_runs"] = current_index_df.runs.values[0] - df2_ball.loc[(df2_ball.id == i) & (df2_ball.inning == inning) & (df2_ball.over < int(str(current_index_df.over.values[0]).split(".")[0]) - 5+1), "runs"].values[0]
                df2_ball.loc[df2_ball.index == index[j], "last_5_over_balls"] = current_index_df.ball.values[0] - df2_ball.loc[(df2_ball.id == i) & (df2_ball.inning == inning) & (df2_ball.over < int(str(current_index_df.over.values[0]).split(".")[0]) - 5+1), "ball"].values[0]
```
