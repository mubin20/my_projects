## Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
pd.pandas.set_option('display.max_columns',None) # Display All Columns of dataframe
```

```python
housing = pd.read_csv('../datasets/train_clean.csv') # Import the clean data file 
```

```python
housing.head()
```

```python
housing.drop(columns='Unnamed: 0',inplace=True) # drop unnamed column
```

```python
housing.head()
```

## Numerical Variables

```python
# List of numerical variables

# 1) List comprehension to filter out data type (object)
numerical_features = [feature for feature in housing.columns if housing[feature].dtypes != 'O'] 

print('Number of numerical variables: ', len(numerical_features))

# visualise the numerical variables
housing[numerical_features].head()
```

# Heatmap to show correlations between all the numerical features.

```python
corr_matrix = housing.corr()[['saleprice']].sort_values('saleprice')
plt.figure(figsize=(8, 12))
sns.heatmap(corr_matrix, annot = True, vmin = -1, vmax = 1, cmap = 'viridis')
plt.title('Correlations Between Numerical Features and Sale Price', fontdict={'fontsize':14})
plt.ylabel('Feature');
```

```python
corr_target = abs(corr_matrix)
salient_num_features = corr_target.loc[corr_target['saleprice'] > 0.5, :]
salient_num_features

# https://towardsdatascience.com/feature-selection-with-pandas-e3690ad8504b
```

### Above are some of the numeric features which are highly correlated with Sale Price.

### Exploring Datetime variables (i.e year_built, year_remod/add,yr_sold) correlation with saleprice

```python
# list of variables that contain year information
year_feature = [feature for feature in numerical_features if 'yr' in feature or 'year' in feature]

year_feature # list of feature with year or yr in it.
```

```python
# Analyse if there is any relationshiop between year sold and salesprice.
housing.groupby('yr_sold')['saleprice'].median().plot()
plt.xlabel('Year Sold')
plt.ylabel('Median House Price')
plt.title("House Price vs YearSold")
```

### From the above graph we can see that as the year increases the price is decreasing which seem to be not true. Lets explore other features.

```python
## Comparing the difference between year_feature with SalePrice

for feature in year_feature:
    if feature!='yr_sold': #if the feature is not equal to yr_sold
        data=housing.copy() # making a copy of dataset()
        
        ## Capture the difference between year variable and year the house was sold for
        data[feature]=data['yr_sold']-data[feature]

        plt.scatter(data[feature],data['saleprice'])
        plt.xlabel(feature)
        plt.ylabel('saleprice')
        plt.show()
```

### 1) For year_built and salesprice price we can observe that the older the house is lesser the value of the house.
### Ex. If the house is like 120-140 year old the price is much lesser.
### 2) For year of remodification we can observe that the price tends to be on higher side for houses which are recently remodified or added.

## Exploring Discrete Variables

```python
# considering all numerical feature if the unique values is less than 25(we set the threshold to 25).
# we consider it as discrete variables. And also feature should not be part of'year_feature','id' and 'pid'

discrete_feature=[feature for feature in numerical_features if len(housing[feature].unique())<25 and feature not in year_feature+['id'] + ['pid']]
print("Discrete Variables Count: {}".format(len(discrete_feature)))
```

```python
discrete_feature # List of discrete_features
```

```python
## Relationship between discrete_feature and saleprice

for feature in discrete_feature:
    data=housing.copy() # make a copy of dataset
    sns.barplot(x=feature,y='saleprice', data=data, ci=False, estimator=np.median)
    plt.xlabel(feature)
    plt.xticks(rotation = 90)
    plt.ylabel('saleprice')
    plt.title(feature)
    plt.show()
```

## Observations: 1) In the above graph, we can see that Building Classes 120 (1-Story Planned Development Unit), 20 (1-Story 1946 and newer), and 60 (2-Story 1946 & Newer) result in significantly increased sale prices. These building classes might be important features for predicting the price of a building in Ames, Iowa.
## 2) For overall_quality vs saleprice we can see there is an exponential rise in price as the overall quality increases. We can observe that it has monotonic relationship.
## 3) From the overall graphs we can say that there seems to be relationship between discrete features and saleprice.

## Exploring Continuous Variables

```python
# considering all numerical feature which are not a part of discrete feature , year_feature, 'id' and 'pid'.

continuous_feature = [feature for feature in numerical_features if feature not in discrete_feature + year_feature + ['id'] +['pid']]
print(f'Continuous feature Count: {len(continuous_feature)}')
```

```python
## Analyse the continuous values by creating histograms to understand the distribution

for feature in continuous_feature:
    data=housing.copy()
    data[feature].hist(bins=25)
    plt.xlabel(feature)
    plt.ylabel("Count")
    plt.title(feature)
    plt.show()
```

## 1) Some of the histograms like (lot_frontage,saleprice) seems to have Gaussian distribution (Normal Distribution).
## 2) Most of the histograms seems to be skewed.

## Transforming numerical_feature using log transformation to see if it makes it normally distributed.

```python
## Using logarithmic transformation


for feature in continuous_feature: # iterate over the continuous_feature list
    data=housing.copy() # make a copy of data before performing logarithmic transformation
    if 0 in data[feature].unique():
        pass
    else:
        data[feature]=np.log(data[feature]) # applying logarithmic transformation on feature
        data['saleprice']=np.log(data['saleprice']) # Applying logarithmic transformation on Saleprice
        plt.scatter(data[feature],data['saleprice']) 
        plt.xlabel(feature)
        plt.ylabel('saleprice')
        plt.title(feature)
        plt.show()
```

## Observation: 1) After applying log normal distribution on features there seem to be monotonic relationship as seen from the above graph.
## 2) It seems to give positive correlation values. 

## Outliers

```python
# Plot a boxplot to see the outliers in feature
for feature in continuous_feature:
    data = housing.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature]=np.log(data[feature])
        data.boxplot(column=feature)
        plt.ylabel(feature)
        plt.title(feature)
        plt.show()
```

## From the above box plot we can observe that are lot of outliers for lot_frontage, lot_area, saleprice respectively.

## Relationship between categorical features and saleprice

```python
# Relationship between categorical features and saleprice
for feature in categorical_features:
    data = housing.copy() # copy the dataset
    sns.barplot(x=feature,y='saleprice', data=data, ci=False, estimator=np.median) # we use median as out categorical data has outliers
    plt.xlabel(feature)
    plt.xticks(rotation = 90)
    plt.ylabel('saleprice')
    plt.title(feature)
    plt.show()
    
```

### Observations: 1) Ms_Zoning: Zoning Classification appears to result in vastly different sale prices. However, two similar levels appear.Grouping properties from Res. Low Density(RL), Floating Villages(FV), and Res. High Density areas (RH)together could be a could feature in predicting price.
### 2) House_style: House style (2.5Fin Two and one-half story: 2nd level finished) ,(2Story Two story) has higher selling price.
### 3) Garage_type: Houses with built-in garages had the highest sale prices. Built-in garages also have living space above them, so in addition to adding value through a garage, they can also add value through more liveable space.
### 4) Street: Houses with paved road results in higher selling price compare to gravel road.
### 5) Land_contour: Properties located on hillside(HLS) appears to be at higher selling price compare to houses located at banked(Bnk) which has least selling price.
### 6) Building type: Single family homes(1Fam) and townhouses(TwnhsE) sells for higher price.
### 7) External_Quality: House with excellent(Ex) quality of material on the exterior of house impacts price.
### 8) Kitchen_Quality: Houses with excellent kitchen quality sells at higher price.
### 9) Sale_type: New homes and Con(Contract 15% Down payment regular terms) seems to sell for most.
