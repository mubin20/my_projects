### DataFrame ( missing_value, its percentage and data type for each column)

```python
# Create a data frame to show columns with missing value and percentage for Train Data.

# 1) sort out null values for each column.
missing_value_train = housing_train.isnull().sum() 

# 2) percentage of missing value for each column.
percent_missing_train = housing_train.isnull().sum() * 100 / len(housing_train) # define

# 3) Data type for column with missing value.
data_type_train = housing_train.dtypes

# 3) Create a dataframe for number of missing values and total percentage per column.
missing_value_df_train = pd.DataFrame({'column_name': housing_train.columns,'missing_value':missing_value_train,
                                 'percent_missing': percent_missing_train,'data_type':data_type_train})

# 4) sort out the dataframe and display only values greater than 0.
missing_value_df_train.sort_values('percent_missing',inplace=True,ascending=False,ignore_index=True)
final_missing_df_train = missing_value_df_train[missing_value_df_train.missing_value > 0]
final_missing_df_train
```

### Dropping columns with  80% and above missing values (pool_qc, misc_feature,alley,fence)

```python
housing_train.drop(columns=['pool_qc','misc_feature','alley','fence'],inplace=True)
```

## Categorical Data
[back to top](#Contents:)
### Fill in all categorical data with mode function.
#### The mode of a set of data values is the value that appears most often.

```python
# Fire place Quality

housing_train['fireplace_qu'] = housing_train['fireplace_qu'].fillna(housing_train['fireplace_qu'].mode()[0])
print(f'The Null value for fireplace quality is : {housing_train["fireplace_qu"].isna().sum()}')
```

```python
# Garage features

housing_train['garage_finish'] = housing_train['garage_finish'].fillna(housing_train['garage_finish'].mode()[0])
print(f'The sum of null value for garage_finish : {housing_train["garage_finish"].isna().sum()}')

housing_train['garage_qual'] = housing_train['garage_qual'].fillna(housing_train['garage_qual'].mode()[0])
print(f'The sum of null value for garage_qual : {housing_train["garage_qual"].isna().sum()}')

housing_train['garage_cond'] = housing_train['garage_cond'].fillna(housing_train['garage_cond'].mode()[0])
print(f'The sum of null value for garage_cond : {housing_train["garage_cond"].isna().sum()}')

housing_train['garage_type'] = housing_train['garage_type'].fillna(housing_train['garage_type'].mode()[0])
print(f'The sum of null value for garage_type : {housing_train["garage_type"].isna().sum()}')
```

```python
# Basement features

housing_train['bsmt_exposure'] = housing_train['bsmt_exposure'].fillna(housing_train['bsmt_exposure'].mode()[0])
print(f'The sum of null value for bsmt_exposure : {housing_train["bsmt_exposure"].isna().sum()}')

housing_train['bsmtfin_type_2'] = housing_train['bsmtfin_type_2'].fillna(housing_train['bsmtfin_type_2'].mode()[0])
print(f'The sum of null value for bsmtfin_type_2 : {housing_train["bsmtfin_type_2"].isna().sum()}')

housing_train['bsmtfin_type_1'] = housing_train['bsmtfin_type_1'].fillna(housing_train['bsmtfin_type_1'].mode()[0])
print(f'The sum of null value for bsmtfin_type_1 : {housing_train["bsmtfin_type_1"].isna().sum()}')

housing_train['bsmt_cond'] = housing_train['bsmt_cond'].fillna(housing_train['bsmt_cond'].mode()[0])
print(f'The sum of null value for bsmt_cond : {housing_train["bsmt_cond"].isna().sum()}')

housing_train['bsmt_qual'] = housing_train['bsmt_qual'].fillna(housing_train['bsmt_qual'].mode()[0])
print(f'The sum of null value for bsmt_qual : {housing_train["bsmt_qual"].isna().sum()}')
```

```python
# Masonry Veneer type

housing_train['mas_vnr_type'] = housing_train['mas_vnr_type'].fillna(housing_train['mas_vnr_type'].mode()[0])
print(f'The sum of null value for mas_vnr_type : {housing_train["mas_vnr_type"].isna().sum()}')
```

```python
sns.heatmap(housing_train.isnull(),yticklabels=False,cbar=False)
```

### Observation: From the above heatmap observation we can see that we manage to fill in all the missing values for categorical data with mode() function.

## Numerical Data
[back to top](#Contents:)
### Fill in all the Numeric data with mean() function. It will fill in all the null values with mean value of that particular column.

```python
# LotFrontage: Linear feet of street connected to property

housing_train['lot_frontage']=housing_train['lot_frontage'].fillna(housing_train['lot_frontage'].mean())
print(f'The sum of null value for lot_frontage : {housing_train["lot_frontage"].isna().sum()}')
```

```python
# MasVnrArea: Masonry veneer area in square feet

housing_train['mas_vnr_area']=housing_train['mas_vnr_area'].fillna(housing_train['mas_vnr_area'].mean())
print(f'The sum of null value for mas_vnr_area : {housing_train["mas_vnr_area"].isna().sum()}')
```

```python
# Basement Features

housing_train['bsmt_half_bath']=housing_train['bsmt_half_bath'].fillna(housing_train['bsmt_half_bath'].mean())
print(f'The sum of null value for bsmt_half_bath : {housing_train["bsmt_half_bath"].isna().sum()}')

housing_train['bsmt_full_bath']=housing_train['bsmt_full_bath'].fillna(housing_train['bsmt_full_bath'].mean())
print(f'The sum of null value for bsmt_full_bath : {housing_train["bsmt_full_bath"].isna().sum()}')

housing_train['bsmt_unf_sf']=housing_train['bsmt_unf_sf'].fillna(housing_train['bsmt_unf_sf'].mean())
print(f'The sum of null value for bsmt_unf_sf : {housing_train["bsmt_unf_sf"].isna().sum()}')

housing_train['total_bsmt_sf']=housing_train['total_bsmt_sf'].fillna(housing_train['total_bsmt_sf'].mean())
print(f'The sum of null value for total_bsmt_sf : {housing_train["total_bsmt_sf"].isna().sum()}')

housing_train['bsmtfin_sf_1']=housing_train['bsmtfin_sf_1'].fillna(housing_train['bsmtfin_sf_1'].mean())
print(f'The sum of null value for bsmtfin_sf_1 : {housing_train["bsmtfin_sf_1"].isna().sum()}')

housing_train['bsmtfin_sf_2']=housing_train['bsmtfin_sf_2'].fillna(housing_train['bsmtfin_sf_2'].mean())
print(f'The sum of null value for bsmtfin_sf_2 : {housing_train["bsmtfin_sf_2"].isna().sum()}')
```

```python
# Garage Features:

housing_train['garage_cars']=housing_train['garage_cars'].fillna(housing_train['garage_cars'].mean())
print(f'The sum of null value for garage_cars : {housing_train["garage_cars"].isna().sum()}')

housing_train['garage_area']=housing_train['garage_area'].fillna(housing_train['garage_area'].mean())
print(f'The sum of null value for garage_area : {housing_train["garage_area"].isna().sum()}')
```

```python
# Function to change data type:
def change_dtype(df, column, map_fxn):
    '''
    Convert a column to a new data type
    '''
    df[column] = df[column].map(map_fxn)
```

```python
# We will convert (id,pid,ms_subclass) to string using above function.
int_to_string = ['id','pid','ms_subclass']

for item in int_to_string:
    change_dtype(housing_train,item,str)
```

```python
housing_train.dtypes
```

```python
# Convert months in integer into month name

months = {1: 'January',
         2: 'February',
         3: 'March',
         4: 'April',
         5: 'May',
         6: 'June',
         7: 'July',
         8: 'August',
         9: 'September',
         10: 'October',
         11: 'November',
         12: 'December'}

change_dtype(housing_train, 'mo_sold', months)
```
