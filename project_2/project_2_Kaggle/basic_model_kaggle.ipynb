# First Submission

```python
features= ['Lot Area','Overall Qual','Overall Cond']
X=train_df[features]
y=train_df['SalePrice']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
lr = LinearRegression()
```

```python
cross_val_score(lr,X_train,y_train).mean()
```

```python
lr.fit(X_train,y_train)
```

```python
lr.score(X_train,y_train)
```

```python
lr.score(X_test,y_test)
```

```python
test_df.head()
```

```python
# now we have to predict the Sales price for the test data
```

```python
features
```

```python
X_pred = test_df[features]
```

```python
X_pred.head()
```

```python
X_pred['future_sale_price'] = lr.predict(X_pred)
```

```python
X_pred.head()
```

```python
X_pred['Id'] = test_df['Id']
```

```python
X_pred.head()
```

```python
X_pred.drop(columns=['Lot Area','Overall Qual','Overall Cond'],inplace=True)
```

```python
X_pred.head()
```

```python
X_pred.set_index('Id',inplace=True)
```

```python
X_pred.head()
```

```python
X_pred.rename(columns={'future_sale_price':'SalePrice'}, inplace=True)
```

```python
X_pred.head()
```

```python
X_pred.to_csv('./first_kaggle_submission.csv')
```
